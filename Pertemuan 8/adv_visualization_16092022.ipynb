# WordCloud Plots

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
# Membaca dan membuka data nya
# open file
alice_novel = open('alice.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
# inisiasi wordcloud
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate wordcloud
alice_wc.generate(alice_novel)
```

```python
# tampilkan wordcloud
import matplotlib.pyplot as plt
plt.imshow(alice_wc, interpolation='bilinear');
plt.axis('off');
```

# Regression Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Canada.xlsx",
                  sheet_name = "Canada by Citizenship",
                  skiprows = 20,
                  skipfooter = 2)
df.head()
```

```python
kol = ['Type','Coverage','AREA','REG','DEV']
df.drop(kol,inplace=True,axis=1)
```

```python
df.rename(columns={'OdName':'Country',
                  'AreaName':'Continent',
                  'RegName':'Region'},inplace=True)
```

```python
df['Total'] = df.sum(axis=1)
```

```python
years = np.arange(1980,2014)
df_total = pd.DataFrame(df[years].sum())
```

```python
df_total.index = map(float,df_total.index)
df_total.reset_index(inplace=True)
df_total.columns = ['years','total']
df_total.head()
```

```python
sns.regplot(x='years',y='total',data=df_total,color='red',marker='*');
```
