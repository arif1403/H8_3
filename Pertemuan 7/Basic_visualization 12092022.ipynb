# Basic Visualization using Python

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Canada.xlsx",
                  sheet_name = "Canada by Citizenship",
                  skiprows = 20,
                  skipfooter = 2)
df.head()
```

```python
kol = ['Type','Coverage','AREA','REG','DEV']
df.drop(kol,inplace=True,axis=1)
```

```python
df.head()
```

```python
df.rename(columns={'OdName':'Country',
                  'AreaName':'Continent',
                  'RegName':'Region'},inplace=True)
```

```python
df.head()
```

```python
df['Total'] = df.sum(axis=1)
```

```python
df.head()
```

```python
df.set_index('Country',inplace=True)
```

```python
df.head()
```

# Ploting

## Line Plot

Line Plot menunjukkan perubahan terhadap waktu

```python
# plot garis untuk data imigran haiti

years = list(range(1980,2014))
haiti = df.loc['Haiti',years]
haiti
```

```python
haiti.plot()
```

```python
# Styling
mpl.style.use(['ggplot'])
```

```python
haiti.plot();
plt.title("Imigrasi from Haiti");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");

plt.text(2000,6400,"Gempa Haiti 2010")
plt.show()
```

**Question 1 :** Bagaimana perbandingan jumlah imigran China dan India dari tahun 1980 - 2013 ke Canada

```python
years = list(range(1980,2014))
negara = ['China','India']
china_india = df.loc[negara,years]
china_india
```

```python
# Transpose
ci_transpose = china_india.T
ci_transpose.head()
```

```python
ci_transpose.plot();
plt.title("Imigran dari China dan India");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
plt.show();
```

**Question 2 :** Bandingkan trend dari negara teratas yang menyumbang imigran ke Canada

```python
top5_index = df['Total'].sort_values(ascending=False).head().index
```

```python
top5_df = df.loc[top5_index,years]
top5_df
```

```python
# Transpose
top5_transpose = top5_df.T
top5_transpose.head()
```

```python
top5_transpose.plot(figsize=(15,8));
plt.title("Imigran 5 teratas");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
plt.show();
```

## Area Plot

```python
top5_transpose.plot(kind='area',
             stacked=False,
             figsize=(15, 8));
plt.title("Imgiran dari 5 negara teratas");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

## Histogram

menggunakan data continue

```python
df_2013 = df[2013]
```

```python
# numpy
count, bin_edges = np.histogram(df_2013, bins=10)

print(count)
print(bin_edges)
```

**Question 3** : Bagaimana distribusi dari Imigran Denmark, Norwegia dan Swedia ?

```python
df_dns = df.loc[["Denmark", "Norway", "Sweden"], years]
df_dns
```

```python
df_dns.T.plot(kind='hist',
              alpha=0.5,
              bins=15,
              figsize=(10,6),
              color=['coral','darkslateblue','mediumseagreen'])
```

## Pie Plot

```python
df_continent = df.groupby('Continent').sum()
df_continent
```

```python
explode_list = [0.1,0,0,0,0.1,0.1]
df_continent['Total'].plot(kind='pie',
                            figsize=(15,6),
                            autopct="%.1f%%",
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.15,
                            explode=explode_list);

plt.legend(labels=df_continent.index);
plt.title("Data Imigrasi sesuai Continent", y=1.12);
plt.axis('equal');
```
