# WordCloud

### Import Library & Load Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
```

```python
alice_novel = open('/SIB_HACKTIVE8/H8_3/Pertemuan 8/alice.txt','r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
# inisiasi wordcloud
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate wordcloud
alice_wc.generate(alice_novel)
```

```python
# tampilkan wordcloud

plt.imshow(alice_wc, interpolation='bilinear');
plt.axis('off');
```
